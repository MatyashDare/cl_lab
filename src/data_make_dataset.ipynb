```python
import json
from datasets import Dataset
def read_data(filepath: str):
    with open(filepath, 'r', encoding='utf-8') as f:
        return json.load(f)
```

```python
root_dir = '/mount/studenten/arbeitsdaten-studenten1/team-lab-phonetics/2023/data/ALC'

wav_dict = {}
num = 0

for folder in tqdm(os.listdir(root_dir)):
    if 'ses' in folder:
        for file in os.listdir(f'{root_dir}/{folder}'):
            if '.json' in file:
                cur_json_path = f'{root_dir}/{folder}/{file}'
                cur_json = read_data(cur_json_path)
                wav_path = f"{folder}/{cur_json['annotates']}"
                
                
                if cur_json['levels'][0]['items'] != []:
                    if num <1000000:
                        wav_dict[num] = {}
                        wav_dict[num]['session'] = folder
                        for label in cur_json['levels'][0]['items'][0]['labels']:
                            
                            wav_dict[num][label['name']] = label['value']
                        num += 1
                    else:
                        break
```
